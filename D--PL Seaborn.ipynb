#### Hist plot

To construct a histogram, follow these steps âˆ’

* Bin the range of values.
* Divide the entire range of values into a series of intervals.
* Count how many values fall into each interval.
  The bins are usually specified as consecutive, non-overlapping intervals of a variable.

* A histogram is a plot that shows the frequency distribution of a set of continuous variable. Histogram 
gives an insight into the underlying distribution (e.g., normal distribution) of the variable, outliers, skewness, etc. To draw a histogram, invoke hist() method of matplotlib library

* In order to draw a histogram in Seaborn, we have a function called distplot and inside that, we have to pass the variable which we want to include. Histogram with default kernel density estimate:


* the histogram is by default include kernel density estimate (kde). Kernel density estimate is the curved line along with the bins or the edges of the frequency of the Ages. If you want to remove the Kernel density estimate (kde) then use kde = False.


* Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. It is used for non-parametric analysis.Setting the hist flag to False in distplot will yield the kernel density estimation plot.distplot() is used to visualize the parametric distribution of a dataset.

* f you want to organize all the different intervals or bins, you can use the bins parameter on the distplot function. Letâ€™s use bins = 5 on the distplot function. It will organize your bins into five bins or intervals.


* Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

# Categorical plot

### BAR PLOT

Bar chart is a frequency chart for qualitative variable (or categorical variable). Bar chart can be used to assess the most-occurring and least-occurring categories within a dataset.

To draw a bar chart, call barplot() of seaborn library.

sns.barplot(x='sex' , y='total_bill', data=tips , ci=0)

We can also create bar charts, which are grouped by a third variable. The parameter hue takes the third variable as parameter. In this case, we pass sex as hue parameter.

sns.barplot(x='sex', y='total_bill', data=tips ,ci=0, hue='sex', palette='bright')

plt.legend(bbox_to_anchor=(1.3,1.0))

# Joint Grid

```python
g=sns.JointGrid(x="total_bill", y="tip", data=tips)
g=g.plot(plt.scatter, sns.distplot)
```

```python
g=sns.JointGrid(x="total_bill", y="tip", data=tips)
g=g.plot(plt.scatter, sns.distplot)
g=g.plot_marginals(sns.distplot,kde=False, color="r")
```

```python
g=sns.JointGrid(x="total_bill", y="tip", data=tips)
g=g.plot(plt.scatter, sns.distplot)
g.ax_marg_x.hist(tips["total_bill"], color="r", alpha=1)
g.ax_marg_y.hist(tips["tip"], color="g", alpha=1, orientation="horizontal")
g.plot_joint(sns.scatterplot, color='black')
```
