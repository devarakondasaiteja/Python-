### MATPLOTLIB

Data Visualization : It is usefull to gain insights and understand what happened in the past in a given context.Itis also helfull for feature engeneering.

Matplotlib library provides simple ways to create boxplots, scatterplots, line graphs, and barcharts distribution plots with custom styling and annotation.Matplotlib is a Python 2D plotting library, Itâ€™s useful for simplifying your data visualization tasks.

Detailed documentation for matplotlib can be found at https://matplotlib.org/.

Seaborn is also a Python data visualization library based on matplotlib. It provides a high-level interface for drawing innovative and informative statistical charts (source: https://seaborn.pydata.org/).

To create graphs and plots, we need to import matplotlib.pyplot and seaborn modules. To display the plots on the Jupyter Notebook, we need to provide a directive %matplotlib inline. Only if the directive is provided, the plots will be displayed on the notebook.

```python
fig=plt.figure()
fig,a=plt.subplots(2,2)
a[0,0].plot(x,y,'y')
a[0,1].plot(y,x,'b')
a[1,0].plot(x,y,'r')
a[1,1].plot(y,x,'g')
```

```python
fig=plt.figure()
fig,a=plt.subplots(1,2)
a[0].plot(x,y)
a[1].plot(y,x)
```

```python
fig=plt.figure()
fig,a=plt.subplots(2,2)

a[0,0].plot(x,y)
a[0,0].set_title('plot 1')

a[0,1].plot(y,x)
a[0,1].set_title('plot 2')

a[1,0].plot(x,y)
a[1,0].set_title('plot 3')

a[1,1].plot(y,x)
a[1,1].set_title('plot 4')

plt.tight_layout();  ## to avoid overlapping(spaces)
plt.tight_layout(pad=3.0);  ## more spaces brtween the plots
```

```python
fig,a=plt.subplots(figsize=(12,3))
a.plot(x,y,'b')
a.set_xlabel('x')
a.set_ylabel('y')
a.set_title('TITLE');
```

```python
# matplotlib style line color and style
fig,ax=plt.subplots()
ax.plot(x,x**2,'b') ## blue line with dots
ax.plot(x,x**3,color='#b22222')  ## http://html color codes.com
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].plot(x,x**2,x,x**3)
axes[0].set_title("Default axes ranges")

axes[1].plot(x,x**2,x,x**3)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5])
axes[1].set_title("custom axes ranges")
plt.suptitle("my plotting");
```
